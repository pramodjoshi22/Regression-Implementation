```python
# importing the libraries
from sklearn import datasets
from sklearn import svm
from sklearn import metrics
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# loading iris data set
iris = datasets.load_iris()
```

```python
print(iris.DESCR)
```

```python
""" 
separating features and labels
iris_x contains features
iris_y contains labels 
"""
iris_X = iris.data
iris_y = iris.target
```

```python
print(iris_X.shape)
```

```python
print(iris_y.shape)
```

```python
# splitting the data set for training and testing parts
indices = np.random.permutation(len(iris_X))

iris_X_train = iris_X[indices[:-20]]
iris_y_train = iris_y[indices[:-20]]
iris_X_test = iris_X[indices[-20:]]
iris_y_test = iris_y[indices[-20:]]
```

```python
print(iris_X_train.shape)
print(iris_X_test.shape)
```

## Hyper-parameter tuning: Finding the best 'C' and 'gamma' values

### Using k = 5 for k-Fold Cross Validation

```python
# creating a KFold object with 5 splits 
folds = KFold(n_splits = 5, shuffle = True, random_state = 101)

# specify range of hyperparameters
c = np.arange(0, 100, 0.25)
gamma = [0.0001, 0.001, 0.01, 0.1]

hyper_params = [{'C': c, 'gamma': gamma}]

# specify model
model = svm.SVC(kernel="rbf")

# set up GridSearchCV()
model_cv = GridSearchCV(estimator = model, param_grid = hyper_params, scoring= 'accuracy', cv = folds, verbose = 1,
                                                                                return_train_score=True, n_jobs = -1)      
# fit the model
model_cv.fit(iris_X_train, iris_y_train)
```

```python
# cv results
cv_results = pd.DataFrame(model_cv.cv_results_)
```

## Plotting the k-Fold Cross Validation results

```python
def get_gamma_results(cv_results, gamma):
    gamma_results = []

    for g in gamma:
        gamma_results.append(cv_results[cv_results['param_gamma']== g])
    
    return gamma_results
```

```python
def plot_results(gamma_rslt, c, gamma, k):
    plt.figure(figsize=(10, 8))

    plt.plot(c, gamma_rslt["mean_test_score"])
    plt.plot(c, gamma_rslt["mean_train_score"])
    
    plt.xlabel('C')
    plt.ylabel('Accuracy')
    title = "Gamma = " + str(gamma) + " and k = " + str(k)
    plt.title(title)
    
    # setting x and y axis range
    plt.yticks(np.arange(0, 1.1, 0.10))
    plt.xticks(np.arange(0, 101, 5))
    
    plt.legend(['test accuracy', 'train accuracy'], loc='best')
```

```python
gamma_results = get_gamma_results(cv_results, gamma)
```

```python
for i in range(len(gamma)):
    plot_results(gamma_results[i], c, gamma[i], 5)
```

## Optimal accuracy score and hyperparameters

```python
# printing the optimal accuracy score and hyperparameters
best_score = model_cv.best_score_
best_hyperparams = model_cv.best_params_

print("The best test score is {0} corresponding to hyperparameters {1} with k = {2}".format(best_score, best_hyperparams, 5))
```

## Training the model using the best parameters and then testing on Unseen dataset 

```python
# rbf kernel using best 'C' and gamma values 
clf = svm.SVC(C = best_hyperparams['C'], gamma = best_hyperparams['gamma'], kernel="rbf")

# fit
clf.fit(iris_X_train, iris_y_train)

# predict
y_pred = clf.predict(iris_X_test)

# accuracy
print("Accuracy:", metrics.accuracy_score(y_true=iris_y_test, y_pred=y_pred), "\n")

# printing the confusion matrix
c_matrix = metrics.confusion_matrix(iris_y_test, y_pred)
print(c_matrix)
```

## Visualization: Effect of gamma on the Decision Boundary of SVM Classifier

```python
def plot_contours(ax, clf, xx, yy, **params):
    """Plot the decision boundaries for a classifier.

    Parameters
    ----------
    ax: matplotlib axes object
    clf: a classifier
    xx: meshgrid ndarray
    yy: meshgrid ndarray
    params: dictionary of params to pass to contourf, optional
    """
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out
```

```python
def make_meshgrid(x, y, h=.02):
    """Create a mesh of points to plot in

    Parameters
    ----------
    x: data to base x-axis meshgrid on
    y: data to base y-axis meshgrid on
    h: stepsize for meshgrid, optional

    Returns
    -------
    xx, yy : ndarray
    """
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    return xx, yy 
```

```python
# drop the last two features for plotting

iris_X_train = iris_X_train[:,:2]
iris_X_test = iris_X_test[:,:2]
```

```python
# plotting decision boundary using gamma, C
def plot_db(C, gamma_, fig_size):
    svc = svm.SVC(kernel="rbf", gamma=gamma_, C=C)
    pred = svc.fit(iris_X_train, iris_y_train)
    U, V = iris_X_train[:, 0], iris_X_train[:, 1]
    xx, yy = make_meshgrid(U, V)
    figsize = fig_size
    fig = plt.figure(figsize=(figsize,figsize))
    ax = plt.subplot(111)
    plot_contours(ax, svc, xx, yy, cmap=plt.cm.coolwarm, alpha=0.8)
    ax.scatter(U, V, c=iris_y_train, cmap=plt.cm.coolwarm, s=20, edgecolors="k")

    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())

    ax.set_xlabel('Sepal length')
    ax.set_ylabel('Sepal width')
    ax.set_xticks(())
    ax.set_yticks(())
    title = "C= " + str(C) + " and gamma= " + str(gamma_)
    ax.set_title(title)

    plt.show()
```
